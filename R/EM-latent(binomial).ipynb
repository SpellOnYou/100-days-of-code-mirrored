### Generate a random sample for the previous two coin experiment. 
Set 80% of the sample as miss_idxing.

See into the data, find MLE: $\hat{\theta_1}, \hat{\theta_2}$ and $\hat{\pi}$ and compute $h_i$ for miss_idxing

```r
set.seed(20181112)
fn2 = function(y, z, params){
  miss_idx = is.na(z)
  for(i in 1:1000L) {
    p = params[1]; th1 = params[2]; th2=params[3]
    p1 <- p * dbinom(y, size=10, prob=th1)
    p2 <- (1-p) * dbinom(y, size=10, prob=th2)
    h1 = p1 / (p1+p2)
    h2 = p2 / (p1+p2)
    z[miss_idx] = h1[miss_idx]
    th1_new = sum(h1*y) / sum(10*h1)
    th2_new = sum((1-h1)*y) / sum(10*(1-h1))
    p_new = mean(z)
    params_new = c(p_new, th1_new, th2_new)
    if(sum(abs(1-params/params_new))<1e-5){
      return(list(theta1=th1_new, theta2 = th2_new, p = p_new, iter=i, h1=h1, h2=h2))
      break}
    
    params = params_new
  }
}
```

```r
n <- 100
ztrue <- rbinom(n, size = 1, prob = 0.5)
z <- ztrue
z[1:80] <- NA

th1 <- 0.2; th2 <- 0.6

y <- ztrue * rbinom(n, size = 10, prob = th1) + (1 - ztrue) * rbinom(n, size = 10, prob = th2)
p=0.5
params = c(p, th1, th2)

fn2(y,z,params)
```
